### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data

```python
df = pd.read_csv("../DATA/AMES_Final_DF.csv")
```

```python
df.head()
```

```python
df.info()
```

**The label we are trying to predict is the SalePrice column. We thus separate out the data into X features and y labels**

```python
X = df.drop('SalePrice',axis=1)
y = df['SalePrice']
```

**We then use scikit-learn to split up X and y into a training set and test set. Since we will later be using a Grid Search strategy, we set our test proportion to 10% and specify random_state = 101**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

**The dataset features has a variety of scales and units. For optimal regression performance, we scale the X features. We fit the scaler only on the training data**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
scaled_X_train = scaler.transform(X_train)
# scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

**We will use an Elastic Net model and thus create an instance of default ElasticNet model with scikit-learn**

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet()
```

**The Elastic Net model has two main parameters, alpha and the L1 ratio. We create a dictionary parameter grid of values for the ElasticNet.**

```python
param_grid = {'alpha':[0.1, 1, 5, 10, 50, 100], 'l1_ratio':[0.01, 0.1, 0.5, 0.75, 0.9, 1]}
```

**Using scikit-learn we create a GridSearchCV object and run a grid search for the best parameters for your model based on our scaled training data.**

```python
### We can select the scoring parameter for the gridsearch model from the below list of scorers
from sklearn.metrics import get_scorer_names

all_scorers = get_scorer_names()
display(all_scorers)
```

```python
from sklearn.model_selection import GridSearchCV

grid_model = GridSearchCV(estimator=model,param_grid=param_grid,scoring='neg_mean_squared_error',cv=10,verbose=1)

grid_model.fit(scaled_X_train, y_train)
```

**We then check the best combination of parameters for our model**

```python
# display(grid_model.best_estimator_)

display(grid_model.best_params_)
```

**We evaluate our model's performance on the unseen 10% scaled test set (hold out set).**

```python
# tuned_model = ElasticNet(alpha=100,l1_ratio=1)
# tuned_model.fit(X_train,y_train)
# y_pred = tuned_model.predict(X_test)
```

```python
y_pred = grid_model.predict(scaled_X_test) ### The grid model can directly be used to generate predictions using the hold out test set
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
display(mae)
# display(mse)
display(rmse)
```
